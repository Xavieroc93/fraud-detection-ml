# 3 Exploraciòn de datos

### Separar los datos de entrada y los datos de salida.

```python
df_label_encoding.head()
```

```python
# Antes de crear X e y, elimina las variables duplicadas/no deseadas
vars_a_eliminar = [
    # 'DE6_Monto_Dolar_log_robust',
    'DE6_Monto_Dolar_log_standard',
    'DE6_Monto_Dolar_log_minmax',
    'DE6_Monto_Dolar_log',
    'DE6_Monto_Dolar'
]

df_label_encoding = df_label_encoding.drop(columns=vars_a_eliminar)
```

```python
# Preparación del dataset final para el modelo
# 'X' contendrá todas las variables independientes (características) del modelo,
# es decir, todos los datos EXCEPTO la variable objetivo ('Fraude').

X = df_label_encoding.drop('Fraude', axis=1)

# 'y' variable objetivo (target) a predecir.
y = df_label_encoding['Fraude']
```

### Entrenamiento y evaluación de modelos.

```python
# División del conjunto de datos en entrenamiento y prueba
# Utiliza la función train_test_split de scikit-learn:
# - 'X' son tus variables independientes, 'y' la variable objetivo
# - test_size=0.2: el 20% de los datos será para prueba, el 80% para entrenamiento
# - random_state=42: asegura que la división sea siempre igual (reproducibilidad) (misma aleatoriedad siempre)
# - stratify=y: garantiza que la proporción de clases en 'y' sea la misma en train y test, evitando desbalance

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=42,stratify=y)

# Resultado: tienes X_train, y_train para entrenar el modelo
# y X_test, y_test para evaluar la calidad y generalización del modelo después de entrenar.
```

### Construir un modelo de regresión con árbol de decisión con Scikit-Learn

- criterion: la función utilizada para medir el error; utilizamos «squared_error».
- max_depth: la profundidad máxima que puede alcanzar el árbol.

```python


# Crea una instancia del modelo de árbol de decisión para regresión, configurando sus parámetros principales:
fraudTree_log_robust = DecisionTreeClassifier(
    criterion='entropy',  # Indica que la función para dividir nodos será el error cuadrático medio (MSE)
    max_depth=8,                # Limita la profundidad máxima del árbol a 8 niveles para controlar el sobreajuste
    random_state=35             # Fija la semilla aleatoria para que los resultados sean reproducibles
)
```

```python
fraudTree_log_robust.fit(X_train, y_train)
```

### Evaluar los modelos de regresión con árbol de decisión de Scikit-Learn y Snap ML
Para evaluar nuestro conjunto de datos, utilizaremos el método de puntuación del objeto DecisionTreeRegressor que proporciona nuestros datos de prueba. Este número es el  valor que indica el coeficiente de determinación. También evaluaremos el error cuadrático medio  de la salida de la regresión con respecto a los valores objetivo del conjunto de prueba. Se esperan valores altos y bajos de un buen modelo de regresión.

- Cuanto más bajo el MSE, mejor el modelo (menor error entre lo predicho y el valor real).
- R^2: Mide la proporción de la varianza total de la variable objetivo que es explicada por el modelo.

```python
# Identificar las 10 características principales con mayor efecto en Fraude

correlation_values = df_label_encoding.corr()['Fraude'].drop('Fraude')
abs(correlation_values).sort_values(ascending=False)[:10]
```

```python
# mostremos la importancia relativa dada a cada característica:
importances = fraudTree_log_robust.feature_importances_
importances
```

```python
# Podemos mostrar estos valores en un gráfico de barras: Para ello extraemos los nombres de las características:
feature_names = X.columns

# y los índices de las importancias dadas, ordenadas de mayor a menor:
indices = np.argsort(importances)
indices
```

```python
# Ahora ordenamos los nombres de las características y los valores obtenidos:
sorted_importances = importances[indices]
sorted_feature_names = feature_names[indices]
```

```python
# Por último, mostramos la gráfica:

fig, ax = plt.subplots(figsize=(18, 8))  # Puedes probar (12, 8) o (14, 10)
ax.barh(y=sorted_feature_names, width=sorted_importances, zorder=10)
ax.grid(color="#EEEEEE", zorder=0)
ax.set_xlabel("Feature importance")
ax.set_ylabel("Feature name")
plt.tight_layout()  # Asegura que las etiquetas no se corten
plt.show()
```

```python
# características que han recibido alguna importancia:
features = list(X.columns[np.nonzero(fraudTree_log_robust.feature_importances_)])
features
```

```python
# Realiza la predicción de las clases (etiquetas) para los datos de prueba
# Utiliza el modelo de árbol de decisión entrenado (drugTree)
# X_testset son las características (variables predictoras) del conjunto de prueba

tree_predictions = fraudTree_log_robust.predict(X_test)
```

```python
# MSE y R2
mse_score = mean_squared_error(y_test, tree_predictions)
print('MSE score: {:.3f}'.format(mse_score))

r2_score = fraudTree_log_robust.score(X_test, y_test)
print('R^2 score: {:.3f}'.format(r2_score))

# Accuracy
accuracy = accuracy_score(y_test, tree_predictions)
print('Accuracy: {:.3f}'.format(accuracy))

# F1 Score
f1 = f1_score(y_test, tree_predictions)
print('F1 Score: {:.3f}'.format(f1))

# Precision y Recall
precision = precision_score(y_test, tree_predictions)
recall = recall_score(y_test, tree_predictions)
print('Precision: {:.3f}'.format(precision))
print('Recall: {:.3f}'.format(recall))

# ROC-AUC (solo si el modelo tiene método predict_proba)
if hasattr(fraudTree_log_robust, "predict_proba"):
    y_pred_prob = fraudTree_log_robust.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_pred_prob)
    print('ROC-AUC: {:.3f}'.format(auc))


# Reporte completo de clasificación
print("\nReporte de clasificación:")
print(classification_report(y_test, tree_predictions, digits=3))
```

```python
from sklearn.metrics import confusion_matrix
# Calcula la matriz de confusión entre las etiquetas verdaderas y las predichas
cm = confusion_matrix(y_test, tree_predictions)

# Imprime la matriz de confusión
print("Matriz de confusión:")
print(cm)
```

```python

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicción")
plt.ylabel("Valor real")
plt.title("Matriz de confusión")
plt.show()
```

```python
# Obtén las probabilidades predichas (necesarias para ROC)
y_pred_prob = fraudTree_log_robust.predict_proba(X_test)[:, 1]

# Calcula los valores para la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# Calcula el área bajo la curva (AUC)
auc = roc_auc_score(y_test, y_pred_prob)

# Crea el gráfico
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC - Detección de Fraude (Árbol de Decisión)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
