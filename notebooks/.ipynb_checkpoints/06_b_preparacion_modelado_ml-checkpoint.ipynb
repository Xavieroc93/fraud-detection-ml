# Analisis de variables no cuantitativas

```python
def analisis_cruce_fraude(df, variable, target='Fraude', cmap='Blues'):
    #Tabla de contingencia
    tabla = pd.crosstab(df[variable], df[target])
    
    #Heatmap
    sns.heatmap(tabla, annot=True, fmt='d', cmap=cmap)
    plt.title(f'Heatmap {variable} vs {target}')
    plt.show()
    
    #Estadísticas chi-cuadrado y Cramér's V
    chi2, p, dof, expected = stats.chi2_contingency(tabla)
    n = tabla.sum().sum()
    min_dim = min(tabla.shape) - 1
    cramers_v = np.sqrt(chi2 / (n * min_dim))
    print(f"Valores {variable} Chi2: {chi2:.2f}, p-value: {p:.4f}, Cramér's V: {cramers_v:.3f}")

# Funciòn 
variables = ['canal', 'entry_mode', 'dia_semana', 'hora', 'DE61_13_Cod_Pais', 'DE18_MCC', 'Comercio_Alias','DE32_Cod_Adquiriente','mes']
for var in variables:
    analisis_cruce_fraude(df_fraude_emisor, var)
```

| **Método**             | **Escala / Transformación**       | **¿Resistente a Outliers?** | **Notas principales**                   |
|------------------------|------------------------------------|-----------------------------|-----------------------------------------|
| **StandardScaler**     | Media=0, Std=1                    | No                          | Supone distribución normal              |
| **MinMaxScaler**       | Min-Max (ej. [0,1])               | No                          | Mantiene outliers; escala fija          |
| **RobustScaler**       | Mediana, Rango IQR                 | Sí                          | Minimiza el impacto de outliers         |
| **PowerTransformer**   | Normalización global (Yeo-Johnson) | Sí                          | Reduce sesgo; homogeneiza varianzas     |
| **QuantileTransformer**| Uniforme o normal por cuantiles    | Sí                          | Compacta colas; iguala distribuciones   |
| **Log/Función manual** | Según transformación (ej. log)     | Parcialmente                | Solo para valores > 0                   |


```python
def escalar_columnas_continuas(df, exclude_cols=None, metodo='standard', plot_histograms=True):
    """
    Estandariza, normaliza o transforma variables numéricas continuas y grafica histogramas.
    
    Parámetros:
        df (pd.DataFrame): DataFrame original.
        exclude_cols (list): Lista de columnas a excluir del escalado (por defecto: ['year', 'mes', 'day', 'hora', 'Fraude']).
        metodo (str): 'standard', 'robust', 'minmax', 'power' o 'quantile'.
        plot_histograms (bool): Si True, genera histogramas de las columnas escaladas.
    
    Retorna:
        df_escalado (pd.DataFrame): DataFrame con las columnas continuas escaladas y el resto igual.
    """
    # Exclusión por defecto
    if exclude_cols is None:
        exclude_cols = ['year', 'mes', 'day', 'hora', 'Fraude']
    
    # Detectar columnas continuas (float64) a escalar
    continuous_columns = [
        col for col in df.select_dtypes(include=['float64']).columns
        if col not in exclude_cols
    ]
    
    # Inicialización del método de escalado/transformación
    if metodo == 'standard':
        scaler = StandardScaler()
    elif metodo == 'robust':
        scaler = RobustScaler()
    elif metodo == 'minmax':
        scaler = MinMaxScaler()
    elif metodo == 'power':
        scaler = PowerTransformer(method='yeo-johnson', standardize=True)
    elif metodo == 'quantile':
        scaler = QuantileTransformer(output_distribution='normal')
    else:
        raise ValueError("Método no reconocido. Usa 'standard', 'robust', 'minmax', 'power' o 'quantile'.")
    
    # Escalado y reconstrucción del DataFrame
    scaled_features = scaler.fit_transform(df[continuous_columns])
    scaled_df = pd.DataFrame(
        scaled_features,
        columns=scaler.get_feature_names_out(continuous_columns),
        index=df.index
    )
    df_escalado = pd.concat([df.drop(columns=continuous_columns), scaled_df], axis=1)
    
    # Visualización opcional de histogramas
    if plot_histograms:
        for col in scaled_df.columns:
            plt.figure(figsize=(6,4))
            plt.hist(scaled_df[col], bins=30, color='steelblue', edgecolor='k', alpha=0.7)
            plt.title(f'Histograma de {col} ({metodo})')
            plt.xlabel(col)
            plt.ylabel('Frecuencia')
            plt.grid(axis='y', alpha=0.2)
            plt.show()
    
    return df_escalado

# Ejemplo de uso de cada método
scaled_data_standard = escalar_columnas_continuas(df_fraude_emisor, metodo='standard')
scaled_data_robust   = escalar_columnas_continuas(df_fraude_emisor, metodo='robust')
scaled_data_minmax   = escalar_columnas_continuas(df_fraude_emisor, metodo='minmax')
scaled_data_power    = escalar_columnas_continuas(df_fraude_emisor, metodo='power')
scaled_data_quantile = escalar_columnas_continuas(df_fraude_emisor, metodo='quantile')

```

- No se obtiene una distribución normal de la variable utilizando StandardScaler, MinMaxScaler ni RobustScaler. Por este motivo, se procede a aplicar una transformación logarítmica sobre los datos con el fin de compactar los outliers y reducir la asimetría. Posteriormente, se vuelven a aplicar las técnicas de escalado para comparar la evolución de la distribución resultante y evaluar cuál método logra una normalización más adecuada para el análisis y modelado.

```python
## No se logra una normalziaciòn deseada, por lo cual se aplica la Transformación logarítmica: np.log(x + 1) a las variables antes de escalar (esto compacta outliers y dispersa el rango central).
df_fraude_emisor['DE6_Monto_Dolar_log'] = np.log(df_fraude_emisor['DE6_Monto_Dolar'] + 1)
```

## Codificación de Variables Categóricas

| **Técnica**           | **Cómo transforma**                                        | **Adecuado para alta cardinalidad** | **¿Preserva orden?** | **Notas principales**                                                                       |
|-----------------------|-----------------------------------------------------------|-------------------------------------|----------------------|--------------------------------------------------------------------------------------------|
| **One Hot Encoding**  | Cada categoría se convierte en columna binaria (0/1)      | No                                  | No                   | Puede aumentar mucho las columnas; ideal para variables con pocos valores únicos.           |
| **Label Encoding**    | Cada valor se transforma en un entero único               | Parcialmente                        | No*                  | Rápido y simple; resultado puede ser arbitrario para variables nominales.                  |
| **Ordinal Encoding**  | Enteros consecutivos representando cada categoría         | Parcialmente                        | Sí                   | Para variables ordinales (con orden lógico). No usar en nominales.                         |
| **Frequency Encoding**| Sustituye categoría por frecuencia relativa o absoluta     | Sí                                  | No                   | Simple, ayuda con alta cardinalidad pero no captura relación con target.                   |
| **Target Encoding**   | Categoría reemplazada por promedio del target             | Sí                                  | No                   | Riesgo de overfitting, usar con validación cruzada. Refleja relación con variable objetivo.|
| **Binary Encoding**   | Asigna valor binario a enteros, luego columnas por bit    | Sí                                  | No                   | Útil en alta cardinalidad, menos columnas que one-hot; información parcialmente preservada.|
| **Hashing Encoding**  | Aplica función hash para asignar columnas                 | Sí                                  | No                   | Rápido. Posibles colisiones; útil para big data y muchas categorías.                       |


One-hot encoding

Se usa One-hot encoding para variables que tienen baja cardinalidad

```python
# Define tu lista de columnas categóricas (o usa las seleccionadas antes) se limino "'rango_monto'" 'franja_horaria'
categorical_columns = [
    'entry_mode', 'canal', 'dia_semana',
    'mes','hora'
]

# Aplica OneHotEncoder (sparse_output=False para obtener array denso)
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_features = encoder.fit_transform(df_fraude_emisor[categorical_columns])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_columns),
                         index=df_fraude_emisor.index)

# Concatena el resultado al DataFrame original EXCLUYENDO las columnas categóricas originales
prepped_data = pd.concat([df_fraude_emisor.drop(columns=categorical_columns), encoded_df], axis=1)
```

```python

prepped_data.head()
```

Se identifica las variables con alta cardinalidad

```python

vars_high_card = ['DE32_Cod_Adquiriente', 'DE42_Cod_Comercio', 'DE61_13_Cod_Pais', 'Comercio_Alias', 'DE18_MCC']
```

Para preparar los datos categóricos antes de entrenar un modelo de machine learning, Se aplican varias técnicas de codificación que transformen las categorías en valores numéricos que los modelos puedan procesar. 

```python
# Label Encoding

# Copia para label encoding
df_label_encoding = prepped_data.copy()
for col in vars_high_card:
    le = LabelEncoder()
    df_label_encoding[f'{col}_LE'] = le.fit_transform(df_label_encoding[col])
    df_label_encoding.drop(columns=col, inplace=True)

df_label_encoding.head()
```

```python
#  Frequency Encoding

# Copia para Frequency Encoding
df_freq_encoding = prepped_data.copy()
for col in vars_high_card:
    freq_map = df_freq_encoding[col].value_counts().to_dict()
    df_freq_encoding[f'{col}_FE'] = df_freq_encoding[col].map(freq_map)
    df_freq_encoding.drop(columns=col, inplace=True)

df_freq_encoding.head()
```

```python
#Target Encoding (media de fraude por categoría)

# Copia para target encoding
df_target_encoding = prepped_data.copy()
for col in vars_high_card:
    # Calcula la media de la variable objetivo (Fraude) por cada categoría
    target_map = df_target_encoding.groupby(col)['Fraude'].transform('mean')
    df_target_encoding[f'{col}_TE'] = target_map
    df_target_encoding.drop(columns=col, inplace=True)
    
df_target_encoding.head()
```

```python
# Binary enconding


# Copia para binary enconding
binary_encoder = ce.BinaryEncoder(cols=vars_high_card)
df_binary_encoding = binary_encoder.fit_transform(prepped_data)
df_binary_encoding.head()
    
df_binary_encoding.head()
```

```python
# Hashing Encoding

# Copia para Hashing Encoding

# Aplica Hashing Encoding en columnas de alta cardinalidad
# n_components define cuántas columnas hash generas
hashing_encoder = ce.HashingEncoder(cols=vars_high_card, n_components=8)
df_hashing_encoding = hashing_encoder.fit_transform(prepped_data)
df_hashing_encoding.head()
```
