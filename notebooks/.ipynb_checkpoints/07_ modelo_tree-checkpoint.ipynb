## Crear conjunto de pruebas

```python
# Preparación del dataset final para el modelo
# 'X' contendrá todas las variables independientes (características) del modelo,
# es decir, todos los datos EXCEPTO la variable objetivo ('Fraude').
X = df_trx_emisor.drop('Fraude', axis =1)

# 'y' variable objetivo (target) a predecir.
y = df_trx_emisor['Fraude']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3 , random_state= 42, stratify = y)
```

```python
print("X de entrenamiento", len(X_train))
print("X de prueba", len(X_test))
print("y de entrenamiento", len(y_train))
print("y de prueba", len(y_test))

print('proporcion fraude datos entrenamiento: ',y_train.value_counts() / len(y_train))
print('proporcion fraude datos test: ',y_test.value_counts() / len(y_test))
print('proporcion fraude datos completos: ',df_trx_emisor['Fraude'].value_counts() / len(df_trx_emisor))
```

Copia del df original de entrenamiento para poder volver a él posteriormente

```python
# Uniendo X_train e y_train en un solo DataFrame
df_train = pd.concat([X_train, y_train], axis=1)
```

# Combinación y creaciòn de atributos

```python
df_train['pais_mcc'] = df_train['DE61_13_Cod_Pais'] + "_"+ df_train['DE18_MCC']
df_train.head()
```

la distribución:
- El primer bin: incluye todos los valores desde 0 hasta 0.99 inclusive (0 ≤ x ≤ 0.99).
- El segundo bin: desde >1 hasta 10 inclusive (0.99 < x ≤ 10).
- El tercero: desde >10 hasta 100 inclusive (10 < x ≤ 100).
- El cuarto: desde >100 hasta 300 inclusive (100 < x ≤ 300).
- El quinto: desde >300 hasta 1000 inclusive (300 < x ≤ 1000).
- El sexto: desde >1000 hasta infinito (1000 < x < ∞).

```python
# Definir rangos de montos por transacciòn
rangos = [0,0.99,10,100,300,1000, float('inf')]
labels = [
    "0-1",          # Validaciones de IIN 
    "1-10",         # Rangos de montos donde tambien se presentan pruebas
    "10-100",       # Pagos moderados
    "100-300",      # Montos medios
    "300-1000",     # Montos altos
    "+1000"         # Montos muy altos
]

# Crear la nueva columna de rangos

df_train['rango_monto'] = pd.cut(
    df_train['DE6_Monto_Dolar'],
    bins=rangos,
    labels = labels,
    right=True, 
    include_lowest= True

)

# Ver conteo por rango
df_train['rango_monto'].value_counts().sort_index()
```

```python
#Crear variables Año, mes,dia, hora, dia semana
#Se generan variables temporales (año, mes, día, hora, trimestre, mes-año) a partir de la fecha de transacció y una variable binaria de fraude.
# df_train['year'] =df_train['DE13_Fecha'].dt.year
df_train['mes'] = df_train['DE13_Fecha'].dt.month
# df_train['day'] = df_train['DE13_Fecha'].dt.day
df_train['hora'] = df_train['DE13_Fecha'].dt.hour
# df_train['mes_year'] = df_train['DE13_Fecha'].dt.to_period('M')
# df_train['trimestre'] = df_train['DE13_Fecha'].dt.to_period('Q')
```

```python

# Define los rangos e intervalos
horarios = [0, 6, 10, 14, 18, 22, 24]
labels = [
    "Madrugada",      # 00:00-05:59
    "Manana",         # 06:00-09:59
    "Medio_dia",      # 10:00-13:59
    "Tarde",          # 14:00-17:59
    "Noche",          # 18:00-21:59
    "Noche_tardia"    # 22:00-23:59
]

# Crear la columna usando pd.cut
df_train['rango_horario'] = pd.cut(
    df_train['hora'],
    bins=horarios,
    labels=labels,
    right=False,
    include_lowest=True
)

# Ver conteo por rango
df_train['rango_horario'].value_counts().sort_index()
```

```python
# Paso 1: Crear la columna de ventana de 15 minutos
df_train['ventana_15min'] = df_train['DE13_Fecha'].dt.floor('15T')

# Paso 2: Ordenar por comercio, ventana y fecha
df_train = df_train.sort_values(['Comercio_Alias', 'ventana_15min', 'DE13_Fecha'])

# Paso 3: Crear el contador acumulativo por comercio y ventana de 15 minutos
df_train['contador_acumulativo_comercio_15min'] = df_train.groupby(['Comercio_Alias', 'ventana_15min']).cumcount() + 1
```
